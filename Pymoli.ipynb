```python
import pandas as pd 
```

```python
df = "./Resources/purchase_data.json"

game_pd=pd.read_json(df)
game_pd.head(20)
```

```python
#list of unique screen names
num_players=list(game_pd["SN"].unique())

num_total=len(num_players)

#Display
pd.DataFrame({"Total Players": [num_total]})
```

```python
#Unique Items
unique_items=list(game_pd["Item ID"].unique())

#Average Price
avg_price=game_pd["Price"].mean()

#Total Purchases
tp=game_pd["Item ID"].count()

#Revenue
rev=game_pd["Price"].sum()

#Display
pd.DataFrame({"Number of unique Items": [len(unique_items)],"Average Price": [avg_price],"Total Purchases":[tp],
             "Revenue":[rev]})

```

```python
#drop repeating same player
player_demographics=game_pd.loc[:,["Gender","SN","Age"]]
player_demographics=player_demographics.drop_duplicates()
gender_breakdown=player_demographics["Gender"].value_counts()

gender_percentage=(gender_breakdown / num_total)*100
gender_demographics = pd.DataFrame({"Total Count": gender_breakdown, "Percentage of Players": gender_percentage})
gender_demographics
```

```python
gender_total_purchase= game_pd.groupby(["Gender"]).sum()["Price"]
gender_mean_purchase=game_pd.groupby(["Gender"]).mean()["Price"]

purchase_data = pd.DataFrame({"Purchase Count": gender_demographics["Total Count"], "Average Purchase Price": gender_mean_purchase, "Total Purchase Value": gender_total_purchase })

purchase_data
```
